```python
import datetime
```

```python
alldata=pd.read_csv("Enigma-JHU.csv", parse_dates=['last_update'])
alldata['last_update']=alldata.last_update.dt.strftime('%Y-%m-%d')
```

```python
countrywise_latest=alldata.query( 'last_update == "2020-04-11" and province_state != province_state and confirmed > 6000' )
countrywise_latest
```

```python
plt.rcParams["figure.figsize"] = (20,12)
plt.bar(countrywise_latest.country_region, countrywise_latest.confirmed)
```
