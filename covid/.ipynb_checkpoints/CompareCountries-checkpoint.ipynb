```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import datetime
```

```python
alldata=pd.read_csv("Enigma-JHU.csv", parse_dates=['last_update'])
```

```python
countries=['India', 'Denmark','South Korea','Saudi Arabia','Singapore']
```

```python
for c in countries:
    print (c)
```

```python
i=0
ad=[]
for c in countries:
    ad.append(alldata.query('combined_key == @c'))
```

```python
for data in ad:
    plt.plot(data.last_update, data.deaths)
plt.legend(countries)
```

```python
for data in ad:
    plt.plot(data.last_update, data.recovered)
plt.legend(countries)
```

```python
for data in ad:
    plt.plot(data.last_update, data.confirmed)
plt.legend(countries)
```
